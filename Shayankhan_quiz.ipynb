## Question 4:

You have a long flowerbed where some plots are already planted with flowers (represented by `1`) and some are empty (represented by `0`). Flowers cannot be planted in adjacent plots.

Given an integer array `flowerbed` where `0` means empty and `1` means occupied, and an integer `n`, determine if you can plant `n` new flowers in the flowerbed without violating the no-adjacent-flowers rule. Return `true` if it's possible, and `false` otherwise.

**Examples:**

1. **Input:** `flowerbed = [1,0,0,0,1]`, `n = 1`  
   **Output:** `true`  
   **Explanation:** You can plant one flower in the middle of the flowerbed, making it `[1,0,1,0,1]`.

2. **Input:** `flowerbed = [1,0,0,0,1]`, `n = 2`  
   **Output:** `false`  
   **Explanation:** It is not possible to plant two flowers without having them next to each other.

**Constraints:**
- `1 <= flowerbed.length <= 20,000`
- `flowerbed[i]` is either `0` or `1`.
- There are no two adjacent flowers in the flowerbed.
- `0 <= n <= flowerbed.length`


````python
def quiz4():
    flowerbed = []
    size = input("Enter the size of flowerbed : ")
    size = int(size)
    n = input("Enter number of n you want to plant : ")
    n = int(n)
    if size <= 1 or size >= 20000 and n <=0 or n >= size:
        print("Invalid size")
        return
    else:
        for i in range(size):
    
            a = input("Enter array value")
            a = int(a)
            if a == 0 or a ==1:
                flowerbed.append(a)
                
            else:
                print("You can enter either 0 or 1")
                return
                
    
    flowerbed = [0] + flowerbed + [0]    
    for i in range(1, size - 1):
        if sum(flowerbed[i - 1 : i + 2]) == 0:
            flowerbed[i] = 1
        n -= 1
        return n <= 0
        
        
quiz4()   
````
